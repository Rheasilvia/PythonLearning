# 统计加州数据

```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_csv('Pandas/state-population.csv')
areas = pd.read_csv('Pandas/state-areas.csv')
abbrevs = pd.read_csv('Pandas/state-abbrevs.csv')

print(pop.head());
print(areas.head());
print(abbrevs.head())
```

```python
merged = pd.merge(pop, abbrevs, how='outer', left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation',1)  #丢弃重复的数据，丢弃整列
merged.head()
```

```python
#检查数据是否缺失
merged.isnull().any()#部分population和state缺失
```

```python
merged[merged['population'].isnull()].head()
```

```python
merged.loc[merged['state'].isnull(),'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()
#state没有缺省值了
```

```python
#共同的state列合并
final = pd.merge(merged, areas, on='state', how='left')
final.head()
```

```python
#再次检查哪列缺失数据
final.isnull().any()
```

```python
#area缺失数据
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
#去掉缺省值，因为全国人口密度再在此无关
final.dropna(inplace=True)
final.head()
```

```python
##用query快速查询计算
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```

```python
#首先对索引进行重置
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']

density.sort_values(ascending=False, inplace=True)
density.head()
```

```python
#计算结果是美国各州加上华盛顿地区...
density.tail()
```
